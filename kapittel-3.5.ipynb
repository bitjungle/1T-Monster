# 3.29-159-466

```python
def f(x):
    return 10 * 0.8**x 

def g(x):
    return (1/2) * 2**x 

def h(t):
    return 1.4 * 1.09**t

def i(x):
    return 10 * 0.75**x 

xverdier = np.linspace(-2, 8, 1000)
y_f = f(xverdier)
y_g = g(xverdier)
y_h = h(xverdier)
y_i = i(xverdier)

fig, ax = plt.subplots(2, 2)
fig.tight_layout()

ax[0, 0].plot(xverdier, y_f)
ax[0, 0].set_title('f(x)')

ax[0, 1].plot(xverdier, y_g)
ax[0, 1].set_title('g(x)')

ax[1, 0].plot(xverdier, y_h)
ax[1, 0].set_title('h(t)')

ax[1, 1].plot(xverdier, y_i)
ax[1, 1].set_title('i(x)')
```

# 3.30-159-466

```python
def O(x):
    return 2 * 1.07**x 

print('a)')
print('Startverdi er 2 mill. og vekstfaktor er (1 + 7/100)')

print('b)')
print('Omsetning etter 4 år er', round(O(4), 4), 'millioner')

print('c)')
print('Lager graf')
aar = np.linspace(0, 20, 21)
oms = O(aar)
plt.plot(aar, oms)
plt.grid()

print('d)')
a = 0
while oms[a] < (2*2): # Leter etter dobling av omsetning
    a += 1 # går ett år framover
print('Dobling av omsetning mellom det', a-1, 'og', a, 'året')
```

# 3.34-160-466

```python
def mikroplast(x):
    return 4000 * 0.94**x

print('a)')
print('4000 * 0.94**x')

print('b)')
print('Om to år:', round(mikroplast(2)), 'tonn')

print('c)')
print('Lager graf')
aar = np.linspace(0, 20, 21)
mengde = mikroplast(aar)
plt.plot(aar, mengde)
plt.grid()

print('d)')
a = 0
while mengde[a] > ((1/2)*4000): # Leter etter halvering av mengde
    a += 1 # går ett år framover
print('Halvering av mengde mellom det', a-1, 'og', a, 'året')
```

# 3.36-160-466

```python
def verdi(aar):
    return 377000 * 0.8 * 0.9**(aar-1)

print('a)')
print('0.8 er verdifall det første året, 0.9**(t-1) er verdifallet de påfølgende årene')

print('b)')
print('Etter ett år:', round(verdi(1)))
print('Etter fem år:', round(verdi(5)))

print('c)')
print('Lager graf')
aar = np.linspace(0, 20, 21)
v = verdi(aar)
plt.plot(aar, v)
plt.grid()
plt.axhline(y=377000/2, color='red')

print('d)')
t = sympy.symbols('t') # Den ukjente i alle likningene nedenfor
likning = sympy.Eq(0.8*0.9**(t-1), 0.5) # Skiller høyre og venstre side i likning med komma
svar = sympy.solve(likning)
print('t=',svar[0]) 
```
