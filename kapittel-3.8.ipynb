```python
# Importerer bibliotek jeg trenger til løsningsforslaga
import math
import matplotlib.pyplot as plt
import numpy as np
import sympy
```

# 3.53-189-467

```python
x = [0,   1,     4,    5,  8, 10]
y = [2, 3.5, 10.25, 12.5, 15, 20]

xverdier = np.linspace(0, 10, 11) # Brukes i oppgave c) og d)

print('a)')
plt.scatter(x, y, marker='x', color='red')

print('b')
plt.plot(x, y, color='cyan')

print('c')
print('Velger de to ytterpunktene, og trekker en linje')
a = (20 - 2) / (10 - 0)
b = 2 - a*0
print('y = ', round(a, 2), 'x +', round(b, 2))
yverdier = a*xverdier + b
plt.plot(xverdier, yverdier, color='blue')


print('d)')
koeff = np.polyfit(x, y, 1)
print('y = ', round(koeff[0], 2), 'x +', round(koeff[1], 2))
yverdier = koeff[0]*xverdier + koeff[1]
plt.plot(xverdier, yverdier, color='magenta')
```

# 3.55-189-467

```python
aar_etter_2010 = [0, 1, 2, 3, 4, 5, 6, 7]
seertid = [110, 110, 96, 93, 88, 71, 70, 49]

print('a)')
koeff = np.polyfit(aar_etter_2010, seertid, 1)
print('y = ', round(koeff[0], 2), 'x +', round(koeff[1], 2))

print('b)')
print('Lager graf')
xverdier = np.linspace(0, 15, 1000)
yverdier = koeff[0]*xverdier + koeff[1]
plt.plot(xverdier, yverdier)
plt.scatter(aar_etter_2010, seertid, color='red')
plt.xlabel('År etter 2010')
plt.ylabel('Seertid')
plt.grid()

print('c)')
print('Det er vanskelig å si noe om gyldigheten til modellen framover i tid.')

print('d)')
i = 0
while yverdier[i] > 0:
    i += 1
print('Slutte å se tv etter', round(xverdier[i], 2), 'år')
```

# 3.56-189-467

```python
aar_etter_1995 = [0, 4, 8, 12, 16, 20]
smakt_alkohol = [79, 85, 84, 77, 70, 57]

print('a)')
koeff = np.polyfit(aar_etter_1995, smakt_alkohol, 2)
print('y = ', round(koeff[0], 2), 'x**2 +', round(koeff[1], 2), 'x +', round(koeff[2], 2))

print('b)')
print('Lager graf')
xverdier = np.linspace(0, 30, 1000)
yverdier = koeff[0]*xverdier**2 + koeff[1]*xverdier + koeff[2]
plt.plot(xverdier, yverdier)
plt.scatter(aar_etter_1995, smakt_alkohol, color='red')
plt.xlabel('År etter 1995')
plt.ylabel('Andel som har smakt alkohol')
plt.grid()

print('c')
x = 2020 - 1995 # år etter 1995
smakt_2020 = koeff[0]*x**2 + koeff[1]*x + koeff[2]
print('Andels som har smakt alkohol i 2020:', round(smakt_2020, 1))

print('d)')
i = 0
while yverdier[i] > 20:
    i += 1
aar = math.floor(1995 + xverdier[i])
print('Andelen vil være 20 % i år', aar)
```

# 3.57-190-467

```python
aar_etter_2010 = [0, 2, 4, 6, 8]
antall = [132, 148, 172, 197, 223]

print('a)')
koeff = np.polyfit(aar_etter_2010, antall, 1)
print('y = ', round(koeff[0], 2), 'x +', round(koeff[1], 2))

print('b)')
print('Kan løses i Python med curve_fit fra scipy.omptimize.')
print('Viser ikke det her, har brukt Geogebra.')
print('y = 131.19 * 1.07**x (fasit er feil)')

print('c)')
print('Lager graf')
xverdier = np.linspace(0, 8, 100)
y_f = koeff[0]*xverdier + koeff[1]
y_g = 131.19 * 1.07**xverdier # Modell laget i Geogebra
plt.plot(xverdier, y_f, label="f(x)")
plt.plot(xverdier, y_g, label="g(x)")
plt.scatter(aar_etter_2010, antall, color="red")
plt.grid()
plt.xlabel('År etter 2010')
plt.ylabel('Antall')

print('d)')
print('Økning på mellom 11 og 12 hytter i året.')

print('e)')
print('Økningen er ca. 7 % per år')

print('f)')
print('Vanskelig å si hvilken modell som er best, men den eksponentielle er best tilpasset.')

print('g)')
x = 2025 - 2010 # år etter 2010
y_f_2025 = koeff[0]*x + koeff[1]
y_g_2025 = 131.19 * 1.07**x # Modell laget i Geogebra
print('f(2025) =', round(y_f_2025), 'g(2025) =', round(y_g_2025))

print('h)')
print('Nei')
```
