# 3.23-154-466

```python
def symmetrilinje(a, b):
    '''Finner symmetrilinje til andregradsfunksjon med 
       koeffisienter a og b'''
    return -b/(2*a)

def ekstremalpunkt(a, b, c):
    '''Finner ekstremalpunkt til andregradsfunksjon med 
       koeffisienter a, b og c'''
    x = symmetrilinje(a, b)
    y = a*x**2 + b*x + c
    return (x, y)

def topp_eller_bunn(a):
    if a < 0:
        return "toppunkt"
    elif a > 0:
        return "bunnpunkt"
    else:
        return None 

print('a)')
print(topp_eller_bunn(2))
print('ekstremalpunkt:', ekstremalpunkt(2, -4, 4))

print('b)')
print(topp_eller_bunn(-1))
print('ekstremalpunkt:', ekstremalpunkt(-1, 0, 9))

print('c)')
print(topp_eller_bunn(1))
print('ekstremalpunkt:', ekstremalpunkt(1, 3, 4))

print('d)')
print(topp_eller_bunn(-1))
print('ekstremalpunkt:', ekstremalpunkt(-1, 2, 0))
```

# 3.24-154-466

```python
def andregrad_nullpunkt(a, b, c):
    '''Finner nullpunkt til andregradsfunksjon med 
       koeffisienter a, b og c'''
    d = b**2 - 4*a*c 
    if d < 0:
        x1 = None
        x2 = None 
    elif d == 0:
        x1 = -b/(2*a)
        x2 = None
    else:
        x1 = (-b + np.sqrt(d)) / (2*a)
        x2 = (-b - np.sqrt(d)) / (2*a)
    return (x1, x2)

print('a)')
print(andregrad_nullpunkt(-1, -2, 3))

print('b)')
print(andregrad_nullpunkt(1, -4, 3))

print('c)')
print(andregrad_nullpunkt(1, 3, 4))

print('d)')
print(andregrad_nullpunkt(1, 0, -25))
```

# 3.25-154-466

```python
def f(x): 
    return x**2 + 2*x - 8

a = 1
b = 2
c = -8

print('a)')
print('Nullpunkter for x =', andregrad_nullpunkt(a, b, c))

print('b)')
print('Symmetrilinje: x =', symmetrilinje(a, b))

print('c)')
e = ekstremalpunkt(a, b, c)
print('Ekstremalpunkt:', e)

print('d)')
minste_f = e[1] # Fant minste funksjonsverdi i forrige oppgave
storste_f = max(f(-6), f(4)) # Sjekker begge endepunkter i definisjonsmengden
print('Vf = [', minste_f, ',', storste_f, ']')

print('e)')
xverdier = np.linspace(-6, 4, 100)
yverdier = f(xverdier)
plt.plot(xverdier, yverdier)
plt.grid()
plt.axhline(y=0, color="black")
plt.axvline(x=0, color="black")
```

# 3.26-155-466

```python
def T(x):
    return 0.08*x**2 - 1.1*x + 19

print('a)')
timer = np.linspace(0, 8, 100)
temperatur = T(timer)
plt.plot(timer, temperatur)
plt.grid()

print('b)')
print('Temperatur ved middnatt var', T(0))

print('c)')
e = ekstremalpunkt(0.08, -1.1, 19)
print('Temperatur på det laveste', e[0], 'timer etter midnatt')

print('d)')
print('Laveste temperatur var', e[1])

print('e)')
print('Vf = [', e[1], ',', T(0), ']')
```

# 3.27-155-466

```python
def f(x):
    return x**3 - x**2 - 5*x + 2

print('a)')
xverdier = np.linspace(-3, 4, 1000)
yverdier = f(xverdier)
plt.plot(xverdier, yverdier)
plt.grid()
plt.axhline(y=0, color="black")
plt.axvline(x=0, color="black")

print('b)')
for i in range(len(yverdier) - 1):
    if yverdier[i] * yverdier[i + 1] < 0:
        xnull = (xverdier[i] + xverdier[i + 1]) / 2
        print('Nullpunkt for x =', round(xnull, 4))

print('c)')
print('Enklest å gjøre ved å derivere, men det har vi ikke gjennomgått ennå :-)')
```
