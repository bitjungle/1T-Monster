# 3.38-166-466

```python
def f(x):
    return 2 * x**1.5

def g(x):
    return x**(2/3)

def h(x):
    return -x**0.8

def i(x):
    return 0.02 * x**2.3

xverdier = np.linspace(0, 5, 100)
y_f = f(xverdier)
y_g = g(xverdier)
y_h = h(xverdier)
y_i = i(xverdier)

fig, ax = plt.subplots(2, 2)
fig.tight_layout()

ax[0, 0].plot(xverdier, y_f)
ax[0, 0].set_title('f(x)')

ax[0, 1].plot(xverdier, y_g)
ax[0, 1].set_title('g(x)')

ax[1, 0].plot(xverdier, y_h)
ax[1, 0].set_title('h(t)')

ax[1, 1].plot(xverdier, y_i)
ax[1, 1].set_title('i(x)')
```

# 3.39-166-467

```python
def f(x):
    return 3.5 * x**1.5

print('a)')
print('f(x) = a * x**b => potensfunksjon')

print('b)')
print('Lager graf')
xverdier = np.linspace(0, 12, 1000)
yverdier = f(xverdier)
plt.plot(xverdier, yverdier)
plt.grid()

print('c)')
print('f(0) =', f(0))
print('f(9) =', f(9))

print('d)')
i = 0
while yverdier[i] < 70.0:
    i += 1
print('f(x) = 70 for x =', xverdier[i])
```

# 3.40-166-467

```python
def V(x):
    return 0.27 * x**1.6

print('a)')
print('Lager graf')
xverdier = np.linspace(0, 20, 1000)
yverdier = V(xverdier)
plt.plot(xverdier, yverdier)
plt.grid()

print('b)')
print('V(8) =', round(V(8), 2))

print('c)')
i = 0
while yverdier[i] < 30.0:
    i += 1
print('V(x) = 30 for x =', round(xverdier[i]))
```

# 3.41-167-467

```python
def H(x):
    return 0.01 * x**2.7

print('a)')
print('H(10) =', round(H(10)))

print('b)')
x = 0
while H(x) < 100:
    x += 0.001 # Små tidsskritt for å unngå å hoppe for langt forbi løsninga
print('H(x) = 100 for x =', round(x, 1))
```

# 3.43-167-467

```python
def L(x):
    return np.cbrt(100*x)

xverdier = np.linspace(0, 3000, 6001)
yverdier = L(xverdier)

print('a)')
print('Lager graf')
plt.plot(xverdier, yverdier)
plt.grid()

print('b)')
print('L(700) =', L(700))

print('c)')
i = 0
while yverdier[i] < 63.0:
    i += 1
print('V(x) = 63 for x =', xverdier[i])
```

# 3.44-167-467

```python
def S(x):
    return np.sqrt(16 - x**2)

xverdier = np.linspace(-4, 4, 1000)
yverdier = S(xverdier)
print('a)')
print('Lager graf')
plt.plot(xverdier, yverdier)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.gca().set_aspect('equal')
plt.axhline(y=0, color='black')
plt.grid()

print('b)')
print('Diameteren er 8. Tallet 16 i funksjonen er radius**2')

print('c)')
print('Kuppelens høyde er radius i halvsirkel, som er 4')

print('d)')
print('sqrt(49 - x**2)')
```
