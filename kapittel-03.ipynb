# 3.9-135-465

```python
def f(x):
    return (-2/3)*x + 4

def finn_nullpunkt(f, x, dx = 0.0001):
    '''Starter ved gitt x, og leter seg fram til første nullpunkt for funksjonen f. 
       OBS: Dersom funksjonen ikke har noe nullpunkt, vil vi ende opp i en evig løkke!
    '''
    while f(x) * f(x + dx) > 0: # Dersom produktet er negativt, har vi et nullpunkt
        x += dx
    return x

print('a)')
print('Når x = 0 er f(x) =', f(0))

print('b)')
x_null = finn_nullpunkt(f, 0)
print('Nullpunkt for x =', round(x_null))
```

# 3.11-135-465

```python
def f_a(x):
    return 3*x + 4

def f_b(x):
    return -6*x + 14

def f_c(x):
    return 2*x 

def f_d(x):
    return 5

print('Skjæring med y-akse (x = 0):')
print('a) x =', f_a(0))
print('b) x =', f_b(0))
print('c) x =', f_c(0))
print('d) x =', f_d(0))

def nullpunkt(f, x_start, x_stopp):
    '''Finner nullpunkt til en funksjon, dersom det er 
       noen nullpunkt mellom x_start og x_stopp.'''
    xverdier = np.linspace(x_start, x_stopp, 1000)
    yverdier = f(xverdier)
    for i in range(0, len(xverdier) - 1):
        if (yverdier[i] * yverdier[i + 1] < 0):
            x_null = (xverdier[i] + xverdier[i + 1]) / 2
            print('Nullpunkt for x = ', x_null)

print('a) Skjæring med x-akse:')
nullpunkt(f_a, -3, 3)

print('b) Skjæring med x-akse:')
nullpunkt(f_b, -3, 3)

print('c) Skjæring med x-akse:')
nullpunkt(f_c, -3, 3)

print('d) Skjæring med x-akse:')
print('Ingen skjæring, funksjonen gir en horisontal linje.')
```

# 3.18-143-465

```python
def linje(punkt_1, punkt_2):
    # a = (y2 - y1) / (x2 - x1)
    a = (punkt_2[1] - punkt_1[1]) / (punkt_2[0] - punkt_1[0]) 
    # b = y1 - a*x1
    b = punkt_1[1] - a * punkt_1[0]
    return {'a': a, "b": b}

print('a)')
print(linje((1, 6), (4, 2)))

print('b)')
print(linje((-4, 1), (2, 3)))

print('c)')
print(linje((0, -2), (5, 2)))

print('d)')
print(linje((-1, -1), (10, -1)))
```

# 3.23-154-466

```python
def symmetrilinje(a, b):
    '''Finner symmetrilinje til andregradsfunksjon med 
       koeffisienter a og b'''
    return -b/(2*a)

def ekstremalpunkt(a, b, c):
    '''Finner ekstremalpunkt til andregradsfunksjon med 
       koeffisienter a, b og c'''
    x = symmetrilinje(a, b)
    y = a*x**2 + b*x + c
    return (x, y)

def topp_eller_bunn(a):
    if a < 0:
        return "toppunkt"
    elif a > 0:
        return "bunnpunkt"
    else:
        return None 

print('a)')
print(topp_eller_bunn(2))
print('ekstremalpunkt:', ekstremalpunkt(2, -4, 4))

print('b)')
print(topp_eller_bunn(-1))
print('ekstremalpunkt:', ekstremalpunkt(-1, 0, 9))

print('c)')
print(topp_eller_bunn(1))
print('ekstremalpunkt:', ekstremalpunkt(1, 3, 4))

print('d)')
print(topp_eller_bunn(-1))
print('ekstremalpunkt:', ekstremalpunkt(-1, 2, 0))
```

# 3.24-154-466

```python
def andregrad_nullpunkt(a, b, c):
    '''Finner nullpunkt til andregradsfunksjon med 
       koeffisienter a, b og c'''
    d = b**2 - 4*a*c 
    if d < 0:
        x1 = None
        x2 = None 
    elif d == 0:
        x1 = -b/(2*a)
        x2 = None
    else:
        x1 = (-b + np.sqrt(d)) / (2*a)
        x2 = (-b - np.sqrt(d)) / (2*a)
    return (x1, x2)

print('a)')
print(andregrad_nullpunkt(-1, -2, 3))

print('b)')
print(andregrad_nullpunkt(1, -4, 3))

print('c)')
print(andregrad_nullpunkt(1, 3, 4))

print('d)')
print(andregrad_nullpunkt(1, 0, -25))
```
