# 3.1-130-465

```python
# Definerer de tre funksjonene som er gitt i oppgavene
def f_a(x):
    return 2*x + 3

def f_b(x):
    return (1/2)*x + 2

def f_c(x):
    return -x + 1

print('a)')
print('f(-1) =', f_a(-1))
print('f(0) =', f_a(0))
print('f(1) =', f_a(1))

print('b)')
print('f(-1) =', f_b(-1))
print('f(0) =', f_b(0))
print('f(1) =', f_b(1))

print('c)')
print('f(-1) =', f_c(-1))
print('f(0) =', f_c(0))
print('f(1) =', f_c(1))
```

# 3.2-130-465

```python
def h(x):
    return 2*x - 3

print('a)')

# Regner ut funksjonsverdien for x = 2
print('h(2) =', h(2))

# Prøver og feiler for å finne y = 5. Vet at x må være større enn 2. 
# Prøver først med x = 3 og fortsetter til funksjonsverdien er 5.
# Legger for sikkerhets skyld inn en stopp for x > 10, sånn at 
# programmet vårt ikke skal gå i en evig løkke om vi har gjort noe feil.
x = 3
while h(x) != 5 and x < 11:
    x += 1

print('x =', x, 'gir h(x) =', h(x))

print('b)')
x = [0, 1, 2, 4] # Fra verditabellen i oppgava
y = [h(i) for i in x] # Regner ut funksjonsverdiene

plt.plot(x, y)
plt.grid()

print('c)')
print('h(5) =', h(5))

print('d)')
plt.axhline(4, color='r') # Tegner en horisontal linje for y = 4
print('Grafisk løsning for y = 4 gir x = 3,5')
```

# 3.3-130-465

```python
def f(x):
    return 10 - 2*x 

print('a)')
print('f(0) =', f(0))
print('f(1) =', f(1))
print('f(2) =', f(2))

print('b)')
print('Funksjonsverdien minker med 2 hver gang x øker med 1')
print('f(3) =', f(3))
print('f(4) =', f(4))

print('c)')
x = [i for i in range(0, 5)]
y = [f(i) for i in x]
print('x:', x)
print('y:', y)

print('d)')
print('f er en funksjon av x fordi det for hver x-verdi finnes nøyaktig en f-verdi')

print('e)')
plt.plot(x, y)
plt.grid()
```

# 3.4-130-465

```python
def f(x):
    return 2 + x**2

print('a)')
print('Vi ser fra grafen at f er en funksjon av x fordi det for hver x-verdi finnes nøyaktig en f-verdi')

print('b)')
print('f(0) =', f(0))

# Prøver og feiler for å finne f(x) = 6. Vet at x må være større enn 1. 
# Prøver først med x = 1 og fortsetter til funksjonsverdien er 6.
# Legger for sikkerhets skyld inn en stopp for x > 10, sånn at 
# programmet vårt ikke skal gå i en evig løkke om vi har gjort noe feil.
x = 1
while f(x) != 6 and x < 11:
    x += 1
print('f(x) = 6 for x = ', x)

print('c)')
# Lager en verditabell med mange x- og y-verdier for å tegne en pen kurve
x = [i/10 for i in range(0, 21)] # Fra 0 til 2 med 0,1 mellom hver verdi
y = [f(i) for i in x]

plt.plot(x, y)
plt.axhline(3, color='r') # Tegner en horisontal linje for y = 3
plt.grid()
print('Grafisk løsning for y = 3 gir x = 1')

print('d)')
print('Definisjonsmengden Df = [0, 2]')
print('Verdimengden Vf = [', f(0), ',', f(2), ']')
```
