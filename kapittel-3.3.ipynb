# 3.14-143-465

```python
def stigningstall(punkt_1, punkt_2):
    # a = (y2 - y1) / (x2 - x1)
    return (punkt_2[1] - punkt_1[1]) / (punkt_2[0] - punkt_1[0])

print('a) Konstantledd er -1, se graf')
print('b) Linja er stigende med økende x-verdier.')
print('c)')
print('Stigningstall a = ', stigningstall((0, -1), (4, 5)))
```

# 3.15-143-465

a) g(x) og h(x)

b) i(x)

c) f(x)

```python
def f(x):
    return [3]*len(x)
def g(x):
    return x - 2
def h(x):
    return x + 3
def i(x):
    return -2*x - 2

xverdier = np.linspace(-4, 6, 3)

plt.plot(xverdier, f(xverdier), label='f(x)')
plt.plot(xverdier, g(xverdier), label='g(x)')
plt.plot(xverdier, h(xverdier), label='h(x)')
plt.plot(xverdier, i(xverdier), label='i(x)')
plt.xlim((-4, 6))
plt.ylim((-4, 6))
plt.grid()
plt.legend()
```

# 3.18-143-465

```python
def linje(punkt_1, punkt_2):
    # a = (y2 - y1) / (x2 - x1)
    a = (punkt_2[1] - punkt_1[1]) / (punkt_2[0] - punkt_1[0]) 
    # b = y1 - a*x1
    b = punkt_1[1] - a * punkt_1[0]
    return {'a': a, "b": b}

print('a)')
print(linje((1, 6), (4, 2)))

print('b)')
print(linje((-4, 1), (2, 3)))

print('c)')
print(linje((0, -2), (5, 2)))

print('d)')
print(linje((-1, -1), (10, -1)))
```
