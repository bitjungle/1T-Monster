# 3.49-174-467

```python
def g(x):
    return 1 / (x + 2)

def vert_asym(c, d):
    ''' f(x) = (ax + b) / (cx + d) '''
    return -(d/c)

def hori_asym(a, c):
    ''' f(x) = (ax + b) / (cx + d) '''
    return (a/c)

print('a)')
print('Nei')

print('b)')
print('Df = R\{-2}')

print('c)')
print('Vertikal asymptote x =', vert_asym(1, 2))

print('d)')
print('Horisontal asymptote y =', hori_asym(0, 1))

print('e)')
xverdier = np.linspace(-4, 0, 500)
yverdier = g(xverdier)

plt.xlim(-4, 0)
plt.ylim(-10, 10)
plt.axvline(x=-2, color="black")
plt.axhline(y=0, color="black")
plt.plot(xverdier, yverdier)
plt.grid()
```

# 3.50-174-467

```python
def f(x):
    return (x - 3)/(3*x - 6)

print('a)')
print('f(100) =', round(f(100), 4))
print('f(1000) =', round(f(1000), 4))
print('Funksjonsverdien går mot 1/3 når x er stor')

print('b)')
print('x =', vert_asym(3, -6))
print('Kan ikke regne ut funksjonsverdien for denne x-verdien')

print('c)')
print('f(1.95) =', f(1.95))
print('f(1.99) =', f(1.99))
print('Funksjonsverdien øker mye når x øker opp mot 2')

print('d)')
print('Vertikal asymptote x =', vert_asym(3, -6))
print('Horisontal asymptote y =', hori_asym(1, 3))

print('e)')
print('Df = R\{2}')
print('Vf = R\{1/3}')

print('f)')
xverdier = np.linspace(-1, 5, 1000)
yverdier = f(xverdier)
plt.xlim(-1, 5)
plt.ylim(-10, 10)
plt.axvline(x=2, color="black")
plt.axhline(y=1/3, color="black")
plt.plot(xverdier, yverdier)
```

# 3.51-174-467

```python
def f(x):
    return (3*x + 4) / (2*x - 4)

print('a)')
print('Nullpunkt =', -4/3)

print('b)')
print('Vertikal asymptote x =', vert_asym(2, -4))
print('Horisontal asymptote y =', hori_asym(3, 2))

print('c)')
print('Df = R\{2}')
print('Vf = R\{3/2}')

print('d)')
xverdier = np.linspace(-1, 5, 1000)
yverdier = f(xverdier)
plt.xlim(-1, 5)
plt.ylim(-20, 20)
plt.axvline(x=2, color="black")
plt.axhline(y=3/2, color="black")
plt.plot(xverdier, yverdier)
```
